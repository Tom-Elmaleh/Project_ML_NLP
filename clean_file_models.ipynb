```python
import pandas as pd
from transformers import pipeline
```

```python
df_test=pd.read_csv('test_set.csv')
df_test.head(5)
df_train=pd.read_csv('train_set.csv')
df_train.head(5)
```

```python
generator = pipeline(model="gpt2")

generated_texts=[]

for element in df_train['avis_en']:

    if len(element) > 1000:
        element = element[:1000]

    generated_texts.append(generator(element, num_return_sequences=1, return_full_text=False)[0]['generated_text'])

df_train['Generation']=generated_texts
```

```python
df_train(5)
```

```python
df_train.to_csv('train_set.csv', index=False)
df_test.to_csv('test_set.csv', index=False)
```
